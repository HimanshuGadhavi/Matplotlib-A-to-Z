### 2D line plot
* Bivariate Analysis
* one column is categorical -> second column is numerical
* or both are numerical data
* Use case - Time series data

```python
# plotting a simple function

price = [48000,54000,57000,49000,47000,45000] #numerical
year = [2015,2016,2017,2018,2019,2020] #categorical

plt.plot(year,price)
plt.show()
```

* if we use any other IDEs and if we do not use the plt.show() function then it will not show the graph so plt.show() is the main function because of which the graph is visible

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
# categorical column = batsman['index']
# numerical column = batsman['V Kohli']
plt.plot(batsman['index'],batsman['V Kohli'])
plt.show()
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.show()
```

```python
# Label title

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit Sharma Vs. Virat Kohli Career Comparision")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.show()
```

>* we can use hex code for different colors from google

```python
# colors(hex) and line(width and style) and marker(size)

plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',
        linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',
        linestyle='dashdot',linewidth=1.8)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',
        linestyle='dashed',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',
        linestyle='dashdot',linewidth=1.8,marker='o',markersize=7)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='green',
        linestyle='dashed',linewidth=3,marker='D',markersize=10,
        label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',
        linestyle='dashdot',linewidth=1.8,marker='o',markersize=7,
        label="Rohit")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.show()
```

*  if we don't call the parameter plt.label() it will not show the output

* suppose some of our values are too large so evantually it will not show the all value in good way so for that we can actually limit the values for any axis

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.show()
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)

plt.xlim(2015,2020)
plt.ylim(0,75000)
plt.show()
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='green',
        linestyle='dashed',linewidth=3,marker='D',markersize=10,
        label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',
        linestyle='dashdot',linewidth=1.8,marker='o',markersize=7,
        label="Rohit")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc="best")
plt.grid() # it will activate the grid
plt.show()
```

### Scatter plot
- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation
![](scatterplot.png)

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
y

plt.scatter(x,y)
plt.show()
```

### Bar charts

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

<img src='bar.png' width='650' height='700' />

```python
# simple bar chart

children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
plt.show()
```

```python
# horizontal bar chart

plt.barh(colors,children,color='black')
plt.show()

'''when number of categories are too large then it is better 
to use the horizontal chart'''
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
# we want bar chart for this data
```

* Multiple Bar charts

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
plt.show()
'''if we try to plot the multiple bar charts by following method then
bars will overlap eachother'''
```

```python
'''in matplotlib bar charts we can control the width of the bar chart'''
plt.bar(df['batsman'],df['2015'],width=0.6)
plt.show()
```

```python
np.arange(df.shape[0])
# here 0 for de villiers 1 for warner and so on 
# we are extracting the index of each player through numpy
```

```python
plt.bar(np.arange(df.shape[0]),df['2015'], width=0.2)
plt.show()
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)

plt.show()
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2,
       color='green')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2,
       color='black')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2,
       color='yellow')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
# xticks modification of labels for each bar
# we will rotate the xticks labels by 60 degree

plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2,
       color='green')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2,
       color='black')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2,
       color='yellow')

plt.xticks(np.arange(df.shape[0]), df['batsman'], rotation=60)
plt.show()
```

* Stacked bar charts
![](stackbar.png)

```python
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])

plt.show()
# here bars will overlap each other
```

```python
# now we will use the parameter bottom
plt.bar(df['batsman'], df['2017'])
plt.bar(df['batsman'], df['2016'], bottom=df['2017'])
plt.bar(df['batsman'], df['2015'], bottom=(df['2016']+df['2017']))

plt.show()
```

```python
plt.bar(df['batsman'], df['2017'], label="2017")
plt.bar(df['batsman'], df['2016'], bottom=df['2017'],
       label="2016")
plt.bar(df['batsman'], df['2015'], bottom=(df['2016']+df['2017']),
       label="2015")

plt.legend(loc="best")
plt.show()
```

### Histogram
- Univariate Analysis
- Numerical col
- Use case - Frequency Count
![](histogram.png)

```python
# simple data
data = [32,45,56,10,15,27,61]

plt.hist(data)
plt.show()
# here bins are made automaticaly according to data
```

```python
# simple data
data = [32,45,56,10,15,27,61]

plt.hist(data,bins=[10,25,40,55,70])
plt.show()
```

### Pie Chart
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale(100%)
![](piechart.svg)

```python
# simple data
data = [23,45,100,20,49]
plt.pie(data)

plt.show()
```

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',
        colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode and shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],
        autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
